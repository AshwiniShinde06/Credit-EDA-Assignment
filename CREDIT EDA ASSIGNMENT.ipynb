# univariate Analysis

```python
%matplotlib inline
```

```python
sns.set_style('darkgrid')
```

```python
def UVAcat(var):
    
    sns.despine
    fig,(x,y) = plt.subplots(1,2,figsize=(30,7))
    
    sns.countplot(x=var, data=App1,ax=x)
    x.set_ylabel('Total Counts')
    x.set_title(f'Distribution of {var} for Non-Defaulters',fontsize=15,color='Blue')
    x.set_xticklabels(x.get_xticklabels(), rotation=60, ha="right")
    
    x.grid(None)
    
    sns.countplot(x=var, data=App2,ax=y)
    y.set_ylabel('Total Counts')
    y.set_title(f'Distribution of {var} for Defaulters',fontsize=15,color='Blue')    
    y.set_xticklabels(y.get_xticklabels(), rotation=60, ha="right")
    
    y.grid(None)

    plt.show()
```

```python
UVAcat('CODE_GENDER')
```

```python
UVAcat('NAME_INCOME_TYPE')
```

```python
UVAcat('FLAG_OWN_CAR')
```

```python
UVAcat('NAME_CONTRACT_TYPE')
```

```python
UVAcat('NAME_FAMILY_STATUS')
```
